```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from IPython.display import display

# 위도와 경도 데이터 전처리 함수
def parse_coordinates(coord_str):
    parts = coord_str.split()
    if len(parts) == 2:
        value, direction = parts
        value = float(value)
        if direction in ['S', 'W']:
            value = -value
        return value
    return None

# 위도와 경도 데이터 전처리
df['위도_숫자'] = df['위도'].apply(parse_coordinates)
df['경도_숫자'] = df['경도'].apply(parse_coordinates)

# 규모를 기준으로 내림차순 정렬하고 상위 20개 선택
top_20_earthquakes = df.sort_values('규모', ascending=False).head(20)

# 한국의 중심 좌표 (대략적인 위치)
korea_center = [36.5, 127.5]

# 지도 생성
m = folium.Map(location=korea_center, zoom_start=7)

# 마커 클러스터 생성
marker_cluster = MarkerCluster().add_to(m)

# 각 지진에 대한 마커 추가
for idx, row in top_20_earthquakes.iterrows():
    if pd.notnull(row['위도_숫자']) and pd.notnull(row['경도_숫자']):
        folium.Marker(
            location=[row['위도_숫자'], row['경도_숫자']],
            popup=f"발생시각: {row['발생시각']}<br>규모: {row['규모']}<br>깊이: {row['깊이(km)']}km<br>진도: {row['진도']}<br>위치: {row['위치']}",
            tooltip=f"규모 {row['규모']} 지진"
        ).add_to(marker_cluster)

# 지도 표시
display(m)

# 결과 출력
print("규모가 큰 상위 20개 지진:")
print(top_20_earthquakes[['발생시각', '규모', '깊이(km)', '진도', '위치', '위도_숫자', '경도_숫자']])

# 기본 통계 정보 출력
print("\n상위 20개 지진의 기본 통계:")
print(top_20_earthquakes['규모'].describe())

# 연도별 상위 20개 지진 분포
year_distribution = top_20_earthquakes['발생시각'].dt.year.value_counts().sort_index()
print("\n연도별 상위 20개 지진 분포:")
print(year_distribution)

# 지역별 상위 20개 지진 분포
location_distribution = top_20_earthquakes['위치'].value_counts()
print("\n지역별 상위 20개 지진 분포:")
print(location_distribution)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# '발생시각' 열을 datetime 형식으로 변환
df['발생시각'] = pd.to_datetime(df['발생시각'], errors='coerce')

# 결측치 제거
df = df.dropna(subset=['발생시각', '규모'])

# 데이터를 시간순으로 정렬
df = df.sort_values('발생시각')

# 그래프 크기 설정
plt.figure(figsize=(14, 8))

# 산점도 및 추세선 그리기
sns.regplot(x=df['발생시각'].astype('int64').astype('float64') / 10**9, 
            y=df['규모'], 
            scatter_kws={'alpha':0.5}, 
            line_kws={'color': 'red'})

# 그래프 꾸미기
plt.title('한국 지진 규모의 시간에 따른 변화')
plt.xlabel('발생 시각')
plt.ylabel('지진 규모')

# x축 날짜 표시 설정
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: pd.to_datetime(x*10**9).strftime('%Y-%m-%d')))
plt.gcf().autofmt_xdate()  # 날짜 레이블 자동 포맷

plt.show()

# 추세선 정보 계산 및 출력
def calculate_trend(df):
    x = df['발생시각'].astype('int64').astype('float64') / 10**9
    y = df['규모']
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    return slope, r_value**2, p_value

slope, r2, p_value = calculate_trend(df)

print("지진 규모 추세:")
print(f"기울기: {slope:.6f}")
print(f"R-squared: {r2:.4f}")
print(f"p-value: {p_value:.4f}")

# 기술 통계량 출력
print("\n지진 규모 통계:")
print(df['규모'].describe())

# 연도별 평균 규모 계산 및 출력
yearly_avg = df.groupby(df['발생시각'].dt.year)['규모'].mean()
print("/n연도별 평균 지진 규모:")
print(yearly_avg)
```

##### 진도가 3이상인 이상치 리스트

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# 이상치 계산 및 출력
Q1 = r['규모'].quantile(0.25)
Q3 = r['규모'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = r[(r['규모'] < lower_bound) | (r['규모'] > upper_bound)]

print("\n이상치 데이터:")
if len(outliers) > 0:
    for index, row in outliers.iterrows():
        print(f"발생시각: {row['발생시각']}, 규모: {row['규모']:.2f}, 위치: {row['위치']}")
else:
    print("이상치가 없습니다.")

# 이상치 개수 출력
print(f"\n이상치 개수: {len(outliers)}")

# 이상치를 제외한 데이터의 통계 정보
normal_data = r[(r['규모'] >= lower_bound) & (r['규모'] <= upper_bound)]
print("\n이상치를 제외한 데이터 통계:")
print(f"데이터 개수: {len(normal_data)}")
print(f"규모 평균: {normal_data['규모'].mean():.2f}")
print(f"규모 중앙값: {normal_data['규모'].median():.2f}")
print(f"규모 최대값: {normal_data['규모'].max():.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# '발생시각' 열을 datetime 형식으로 변환
df['발생시각'] = pd.to_datetime(df['발생시각'], errors='coerce')

# 2018년 7월 15일 이후의 데이터만 선택
df_filtered = df[df['발생시각'] > '2018-07-15']

# 필요한 열만 선택
columns_of_interest = ['규모', '깊이(km)', '진도']
df_correlation = df_filtered[columns_of_interest]

# '-' 값을 NaN으로 변환
df_correlation = df_correlation.replace('-', np.nan)

# 데이터 타입 변환
df_correlation['규모'] = pd.to_numeric(df_correlation['규모'], errors='coerce')
df_correlation['깊이(km)'] = pd.to_numeric(df_correlation['깊이(km)'], errors='coerce')
df_correlation['진도'] = pd.to_numeric(df_correlation['진도'], errors='coerce')

# 결측치 제거
df_correlation = df_correlation.dropna()

# 상관계수 계산
correlation_matrix = df_correlation.corr()

# 상관계수 출력
print("상관계수 행렬:")
print(correlation_matrix)

# 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('규모, 깊이(km), 진도 간의 상관관계 히트맵')
plt.show()

# 산점도 행렬 시각화
sns.pairplot(df_correlation)
plt.suptitle('규모, 깊이(km), 진도 간의 산점도 행렬', y=1.02)
plt.show()

# 각 변수의 기술 통계량 출력
print("\n기술 통계량:")
print(df_correlation.describe())

# 데이터 개수 출력
print("\n데이터 개수:")
print(df_correlation.count())
```

##### 국내 규모 3이상의 지진 추세

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# '발생시각' 열을 datetime 형식으로 변환
df['발생시각'] = pd.to_datetime(df['발생시각'])

# 규모 3 이상의 지진 데이터만 필터링
df_filtered = df[df['규모'] >= 3.0]

# 결측치 제거
df_filtered = df_filtered.dropna(subset=['발생시각', '규모'])

# 데이터를 시간순으로 정렬
df_filtered = df_filtered.sort_values('발생시각')

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 산점도 그리기
plt.scatter(df_filtered['발생시각'], df_filtered['규모'], alpha=0.6)

# 추세선 추가
x = df_filtered['발생시각'].astype('int64') // 10**9  # datetime을 Unix timestamp로 변환
y = df_filtered['규모']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept
plt.plot(df_filtered['발생시각'], line, color='red', label=f'추세선 (R² = {r_value**2:.3f})')

# 그래프 꾸미기
plt.title('시간에 따른 규모 3 이상 지진의 규모 변화')
plt.xlabel('발생 시각')
plt.ylabel('지진 규모')
plt.legend()

# y축 범위 설정 (규모 3부터 시작)
plt.ylim(bottom=3)

# x축 날짜 포맷 설정
plt.gcf().autofmt_xdate()

# 그래프 표시
plt.show()

# 추가적인 통계 정보 출력
print(f"기울기: {slope:.6f}")
print(f"y 절편: {intercept:.6f}")
print(f"R-squared: {r_value**2:.3f}")
print(f"p-value: {p_value:.6f}")

# 데이터 요약 정보
print("\n데이터 요약:")
print(f"규모 3 이상 지진 총 발생 횟수: {len(df_filtered)}")
print(f"첫 번째 지진 발생 시각: {df_filtered['발생시각'].min()}")
print(f"마지막 지진 발생 시각: {df_filtered['발생시각'].max()}")
print(f"평균 규모: {df_filtered['규모'].mean():.2f}")
print(f"최대 규모: {df_filtered['규모'].max():.2f}")
print(f"최소 규모: {df_filtered['규모'].min():.2f}")
```
