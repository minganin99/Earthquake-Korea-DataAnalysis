##### 한국과 오클라호마 지진의 빈도는 연도별로 어떤 추세인가?

```python
# origintime 컬럼을 datetime 자료형으로 변환
df_okc['origintime'] = pd.to_datetime(df_okc['origintime'], errors='coerce')

# 연도 추출
df_okc['연도'] = df_okc['origintime'].dt.year

# 연도별 지진 빈도수 계산
earthquake_count = df_okc['연도'].value_counts().sort_index().reset_index()
earthquake_count.columns = ['연도', '지진빈도수']

# 2000년 이후 데이터 필터링
earthquake_count = earthquake_count[earthquake_count['연도'] >= 2000]

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(data=earthquake_count, x='연도', y='지진빈도수', palette='viridis')
plt.title('2000년 이후 연도별 지진 빈도수')
plt.xlabel('연도')
plt.ylabel('지진 빈도수')
plt.xticks(rotation=45)  # x축 레이블 회전
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# origintime 컬럼을 datetime 자료형으로 변환
df_kor['발생시각'] = pd.to_datetime(df_kor['발생시각'], errors='coerce')

# 연도 추출
df_kor['연도'] = df_kor['발생시각'].dt.year

# 연도별 지진 빈도수 계산
earthquake_count = df_kor['연도'].value_counts().sort_index().reset_index()
earthquake_count.columns = ['연도', '지진빈도수']

# 2000년 이후 데이터 필터링
earthquake_count = earthquake_count[earthquake_count['연도'] >= 2000]

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(data=earthquake_count, x='연도', y='지진빈도수', palette='viridis')
plt.title('2000년 이후 연도별 지진 빈도수')
plt.xlabel('연도')
plt.ylabel('지진 빈도수')
plt.xticks(rotation=45)  # x축 레이블 회전
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# 한국 지진 데이터 전처리
df_kor['발생시각'] = pd.to_datetime(df_kor['발생시각'], errors='coerce')
df_kor['연도'] = df_kor['발생시각'].dt.year
earthquake_count_kor = df_kor['연도'].value_counts().sort_index().reset_index()
earthquake_count_kor.columns = ['연도', '한국 지진 빈도수']
earthquake_count_kor = earthquake_count_kor[earthquake_count_kor['연도'] >= 2000]

# 다른 지역 지진 데이터 전처리
df_okc['origintime'] = pd.to_datetime(df_okc['origintime'], errors='coerce')
df_okc['연도'] = df_okc['origintime'].dt.year
earthquake_count_okc = df_okc['연도'].value_counts().sort_index().reset_index()
earthquake_count_okc.columns = ['연도', '오클라호마 지진 빈도수']
earthquake_count_okc = earthquake_count_okc[earthquake_count_okc['연도'] >= 2000]

# 두 데이터프레임 병합
merged_counts = pd.merge(earthquake_count_kor, earthquake_count_okc, on='연도', how='outer').fillna(0)

# 그래프 그리기
plt.figure(figsize=(12, 6))
sns.lineplot(data=merged_counts, x='연도', y='한국 지진 빈도수', label='한국 지진 빈도수', marker='o')
sns.lineplot(data=merged_counts, x='연도', y='오클라호마 지진 빈도수', label='오클라호마 지진 빈도수', marker='o')
plt.title('2000년 이후 연도별 지진 빈도수 비교')
plt.xlabel('연도')
plt.ylabel('지진 빈도수')
plt.xticks(rotation=45)  # x축 레이블 회전
plt.grid(axis='y')
plt.legend()
plt.tight_layout()
plt.show()
```

##### 한국과 오클라호마의 최대 지진 규모는 언제 발생했나?
- 지진이 많이 발생하기 시작한 기간내에 발생

```python
max_okc = df_okc[df_okc['magnitude'] == df_okc['magnitude'].max()]
max_kor = df_kor[df_kor['규모'] == df_kor['규모'].max()]
print('한국 지진 최고 규모:', max_kor)
print('오클라호마 지진 최고 규모:', max_okc)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df_kor['규모'], bins=30, color='blue', kde=True, label='한국 지진', stat='density')
sns.histplot(df_okc['magnitude'], bins=30, color='orange', kde=True, label='오클라호마 지진', stat='density')
plt.title('지진 규모 분포 비교')
plt.xlabel('규모')
plt.ylabel('밀도')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_kor, x='깊이(km)', y='규모', label='한국 지진', color='blue')
sns.scatterplot(data=df_okc, x='depth_km', y='magnitude', label='오클라호마 지진', color='orange')
plt.title('깊이와 규모의 관계')
plt.xlabel('깊이 (km)')
plt.ylabel('규모')
plt.legend()
plt.grid()
plt.show()
```

```python
# 날짜 변환
df_kor['발생시각'] = pd.to_datetime(df_kor['발생시각'], errors='coerce')
df_okc['origintime'] = pd.to_datetime(df_okc['origintime'], errors='coerce')

# 2000년 이후 데이터 필터링
df_kor_filtered = df_kor[df_kor['발생시각'].dt.year >= 2000]
df_okc_filtered = df_okc[df_okc['origintime'].dt.year >= 2000]

# 7일 이동 평균 계산
earthquake_count_kor = df_kor_filtered['발생시각'].dt.date.value_counts().sort_index().rolling(window=7).mean()
earthquake_count_okc = df_okc_filtered['origintime'].dt.date.value_counts().sort_index().rolling(window=7).mean()

# 시각화
plt.figure(figsize=(12, 6))
sns.lineplot(x=earthquake_count_kor.index, y=earthquake_count_kor.values, label='한국 지진 (7일 이동 평균)', color='blue')
sns.lineplot(x=earthquake_count_okc.index, y=earthquake_count_okc.values, label='오클라호마 지진 (7일 이동 평균)', color='orange')
plt.title('2000년 이후 지진 발생 트렌드 (7일 이동 평균)')
plt.xlabel('날짜')
plt.ylabel('지진 발생 수')
plt.legend()
plt.grid()
plt.show()
```

```python
# 상관 관계 분석
correlation_kor = df_kor[['규모', '깊이(km)']].corr()
correlation_okc = df_okc[['magnitude', 'depth_km']].corr()

plt.figure(figsize=(12, 6))
sns.heatmap(correlation_kor, annot=True, cmap='coolwarm', center=0)
plt.title('한국 지진 변수 간 상관 관계')
plt.show()

plt.figure(figsize=(12, 6))
sns.heatmap(correlation_okc, annot=True, cmap='coolwarm', center=0)
plt.title('오클라호마 지진 변수 간 상관 관계')
plt.show()
```
