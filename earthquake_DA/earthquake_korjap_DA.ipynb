```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import platform

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지
# 데이터 준비 (이미 df_abr_countries가 준비되어 있다고 가정)
df_abr_countries = df_abroad[df_abroad['위치'].str.contains('일본')]

# '발생시각' 열을 datetime 형식으로 변환
df_abr_countries['발생시각'] = pd.to_datetime(df_abr_countries['발생시각'])

# 결측치 제거
df_filtered = df_abr_countries.dropna(subset=['발생시각', '규모'])

# 데이터를 시간순으로 정렬
df_filtered = df_filtered.sort_values('발생시각')

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 산점도 그리기
plt.scatter(df_filtered['발생시각'], df_filtered['규모'], alpha=0.6)

# 추세선 추가
x = (df_filtered['발생시각'] - df_filtered['발생시각'].min()).dt.total_seconds()
y = df_filtered['규모']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept
plt.plot(df_filtered['발생시각'], line, color='red', label=f'추세선 (R² = {r_value**2:.3f})')

# 그래프 꾸미기
plt.title('일본에서 발생한 지진의 시간에 따른 규모 변화')
plt.xlabel('발생 시각')
plt.ylabel('지진 규모')
plt.legend()

# x축 날짜 포맷 설정
plt.gcf().autofmt_xdate()

# 그래프 표시
plt.show()

# 추가적인 통계 정보 출력
print(f"기울기: {slope:.6f}")
print(f"y 절편: {intercept:.6f}")
print(f"R-squared: {r_value**2:.3f}")
print(f"p-value: {p_value:.6f}")

# 데이터 요약 정보
print("\n데이터 요약:")
print(f"일본에서 발생한 지진 총 발생 횟수: {len(df_filtered)}")
print(f"첫 번째 지진 발생 시각: {df_filtered['발생시각'].min()}")
print(f"마지막 지진 발생 시각: {df_filtered['발생시각'].max()}")
print(f"평균 규모: {df_filtered['규모'].mean():.2f}")
print(f"최대 규모: {df_filtered['규모'].max():.2f}")
print(f"최소 규모: {df_filtered['규모'].min():.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지
# 데이터 로드
df_japan = pd.read_csv('시계열_일본_지진데이터.csv')
df_korea = pd.read_excel('C:/Users/campus4D036/Desktop/프로젝트/팀프로젝트/Earthquake-Korea-DataAnalysis/data_set/최종데이터셋/지진/시계열_한국_지진데이터.xlsx')

# '발생시각' 열을 datetime 형식으로 변환
df_japan['발생시각'] = pd.to_datetime(df_japan['발생시각'], errors='coerce')
df_korea['발생시각'] = pd.to_datetime(df_korea['발생시각'], errors='coerce')

# 유효한 날짜 범위만 선택 (예: 1900년부터 현재까지)
df_japan = df_japan[(df_japan['발생시각'].dt.year >= 1900) & (df_japan['발생시각'].dt.year <= 2023)]
df_korea = df_korea[(df_korea['발생시각'].dt.year >= 1900) & (df_korea['발생시각'].dt.year <= 2023)]

# 그래프 크기 설정
plt.figure(figsize=(14, 8))

# 일본 데이터 플롯 및 추세선
sns.regplot(x=df_japan['발생시각'].astype('int64').astype('float64') / 10**9, 
            y=df_japan['규모'], 
            scatter_kws={'alpha':0.3}, 
            line_kws={'color': 'red'}, 
            label='일본')

# 한국 데이터 플롯 및 추세선
sns.regplot(x=df_korea['발생시각'].astype('int64').astype('float64') / 10**9, 
            y=df_korea['규모'], 
            scatter_kws={'alpha':0.3}, 
            line_kws={'color': 'blue'}, 
            label='한국')

# 그래프 꾸미기
plt.title('일본과 한국의 지진 규모 추세 비교')
plt.xlabel('시간')
plt.ylabel('지진 규모')
plt.legend()

# x축 날짜 표시 설정
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: pd.to_datetime(x*10**9).strftime('%Y-%m-%d')))
plt.gcf().autofmt_xdate()  # 날짜 레이블 자동 포맷

plt.show()

# 각 국가별 추세선 정보 계산 및 출력
def calculate_trend(df):
    x = df['발생시각'].astype('int64').astype('float64') / 10**9
    y = df['규모']
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    return slope, r_value**2, p_value

japan_slope, japan_r2, japan_p = calculate_trend(df_japan)
korea_slope, korea_r2, korea_p = calculate_trend(df_korea)

print("일본 지진 규모 추세:")
print(f"기울기: {japan_slope:.6f}")
print(f"R-squared: {japan_r2:.4f}")
print(f"p-value: {japan_p:.4f}")

print("\n한국 지진 규모 추세:")
print(f"기울기: {korea_slope:.6f}")
print(f"R-squared: {korea_r2:.4f}")
print(f"p-value: {korea_p:.4f}")

# 기술 통계량 출력
print("\n일본 지진 규모 통계:")
print(df_japan['규모'].describe())

print("\n한국 지진 규모 통계:")
print(df_korea['규모'].describe())
```

##### 일본에서 발생한 지진 중 한국에 영향을 미칠만 한 지진

```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster
from branca.colormap import LinearColormap

def parse_coordinates(coord_str):
    if pd.isna(coord_str):
        return None
    parts = str(coord_str).split()
    if len(parts) == 2:
        value, direction = parts
        try:
            value = float(value)
            if direction in ['S', 'W']:
                value = -value
            return value
        except ValueError:
            return None
    return None

df_japan['위도_숫자'] = df_japan['위도'].apply(parse_coordinates)
df_japan['경도_숫자'] = df_japan['경도'].apply(parse_coordinates)
df_japan['규모'] = pd.to_numeric(df_japan['규모'], errors='coerce')
df_japan['깊이(km)'] = pd.to_numeric(df_japan['깊이(km)'], errors='coerce')

# 유효한 데이터만 선택 (위도, 경도, 규모가 모두 있는 경우)
valid_earthquakes = df_japan.dropna(subset=['위도_숫자', '경도_숫자', '규모'])

# 지도 생성
m = folium.Map(location=[36, 138], zoom_start=5)

# 색상 맵 생성
colormap = LinearColormap(colors=['blue', 'green', 'yellow', 'orange', 'red'], 
                          vmin=valid_earthquakes['규모'].min(), 
                          vmax=valid_earthquakes['규모'].max())

# 마커 클러스터 생성
marker_cluster = MarkerCluster().add_to(m)

# 각 지진에 대한 마커 추가
for idx, row in valid_earthquakes.iterrows():
    folium.CircleMarker(
        location=[row['위도_숫자'], row['경도_숫자']],
        radius=row['규모'] * 1.5,  # 규모에 따라 원의 크기 조절
        popup=f"발생시각: {row['발생시각']}<br>규모: {row['규모']:.1f}<br>깊이: {row['깊이(km)']}km<br>위치: {row['위치']}",
        tooltip=f"규모 {row['규모']:.1f} 지진",
        color=colormap(row['규모']),
        fill=True,
        fillColor=colormap(row['규모']),
        fillOpacity=0.7
    ).add_to(marker_cluster)

# 색상 범례 추가
colormap.add_to(m)

# 한국의 대략적인 위치에 마커 추가
folium.Marker(
    location=[37.5665, 126.9780],
    popup="Seoul, South Korea",
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(m)

# 지도 표시
display(m)

# 기본 통계 정보 출력
print("\n지진의 기본 통계:")
print(valid_earthquakes['규모'].describe())

# 연도별 지진 분포
valid_earthquakes['발생시각'] = pd.to_datetime(valid_earthquakes['발생시각'])
year_distribution = valid_earthquakes['발생시각'].dt.year.value_counts().sort_index()
print("\n연도별 지진 분포:")
print(year_distribution)

# 규모별 지진 분포
magnitude_distribution = pd.cut(valid_earthquakes['규모'], bins=[0, 3, 4, 5, 6, 7, np.inf], labels=['0-3', '3-4', '4-5', '5-6', '6-7', '7+'])
print("\n규모별 지진 분포:")
print(magnitude_distribution.value_counts().sort_index())
```

```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster


def parse_coordinates(coord_str):
    if pd.isna(coord_str):
        return None
    parts = str(coord_str).split()
    if len(parts) == 2:
        value, direction = parts
        try:
            value = float(value)
            if direction in ['S', 'W']:
                value = -value
            return value
        except ValueError:
            return None
    return None

df_japan['위도_숫자'] = df_japan['위도'].apply(parse_coordinates)
df_japan['경도_숫자'] = df_japan['경도'].apply(parse_coordinates)
df_japan['규모'] = pd.to_numeric(df_japan['규모'], errors='coerce')
df_japan['깊이(km)'] = pd.to_numeric(df_japan['깊이(km)'], errors='coerce')

# 한국에 영향을 줄 만한 지진 선별 (규모 5.0 이상)
relevant_earthquakes = df_japan[
    (df_japan['위도_숫자'] >= 32) & (df_japan['위도_숫자'] <= 41) &
    (df_japan['경도_숫자'] >= 129) & (df_japan['경도_숫자'] <= 140) &
    (df_japan['규모'] >= 5.0) &
    (df_japan['깊이(km)'] <= 100)
]

# 결과 출력
print("한국에 영향을 줄 만한 일본 지진 데이터:")
print(relevant_earthquakes[['발생시각', '규모', '깊이(km)', '위도_숫자', '경도_숫자', '위치']])

# 지도 생성
m = folium.Map(location=[36, 135], zoom_start=5)

# 마커 클러스터 생성
marker_cluster = MarkerCluster().add_to(m)

# 각 지진에 대한 마커 추가
for idx, row in relevant_earthquakes.iterrows():
    folium.CircleMarker(
        location=[row['위도_숫자'], row['경도_숫자']],
        radius=row['규모'] * 1.5,  # 규모에 따라 원의 크기 조절
        popup=f"발생시각: {row['발생시각']}<br>규모: {row['규모']:.1f}<br>깊이: {row['깊이(km)']}km<br>위치: {row['위치']}",
        tooltip=f"규모 {row['규모']:.1f} 지진",
        color='red',
        fill=True,
        fillColor='red'
    ).add_to(marker_cluster)

# 한국의 대략적인 위치에 마커 추가
folium.Marker(
    location=[37.5665, 126.9780],
    popup="Seoul, South Korea",
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(m)

# 지도 표시
display(m)

# 기본 통계 정보 출력
print("\n선별된 지진의 기본 통계:")
print(relevant_earthquakes['규모'].describe())

# 연도별 선별된 지진 분포
relevant_earthquakes['발생시각'] = pd.to_datetime(relevant_earthquakes['발생시각'])
year_distribution = relevant_earthquakes['발생시각'].dt.year.value_counts().sort_index()
print("\n연도별 선별된 지진 분포:")
print(year_distribution)
```
