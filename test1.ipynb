```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager

# 한글 폰트 설정 (Windows 환경)
font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = font_manager.FontProperties(fname=font_path)

# 데이터 로드
df = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/원자력발전소/국내원자력발전소현황.csv', encoding='EUC-KR')

# 날짜 컬럼을 datetime 형식으로 변환
df['상업운전개시일'] = pd.to_datetime(df['상업운전개시일'], errors='coerce')
df['설계수명만료일'] = pd.to_datetime(df['설계수명만료일'], errors='coerce')

# 연도만 추출하여 새로운 컬럼 추가
df['상업운전개시연도'] = df['상업운전개시일'].dt.year
df['설계수명만료연도'] = df['설계수명만료일'].dt.year

# 각 발전소별로 상업운전개시연도의 최솟값과 설계수명만료연도의 최댓값을 구함
df_range = df.groupby('발전소명').agg({'상업운전개시연도': 'min', '설계수명만료연도': 'max'}).reset_index()

# 그래프 시각화
plt.figure(figsize=(14, 8))

# 막대 그래프 생성: 각 발전소의 상업운전개시연도에서 설계수명만료연도까지를 하나의 막대로 표현
for i, row in df_range.iterrows():
    plt.barh(i, row['설계수명만료연도'] - row['상업운전개시연도'], left=row['상업운전개시연도'], color='skyblue', height=0.8)

# 발전소 이름을 y축에 추가
plt.yticks(range(len(df_range)), df_range['발전소명'], fontproperties=font_prop)
plt.xticks(fontsize=12, fontproperties=font_prop)

# 제목 및 레이블 설정
plt.title('발전소별 상업운전개시연도에서 설계수명만료연도까지의 기간 (평균)', fontsize=16, fontproperties=font_prop)
plt.xlabel('연도', fontsize=14, fontproperties=font_prop)
plt.ylabel('발전소명', fontsize=14, fontproperties=font_prop)

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
df_shelter = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/대피소/국내대피소_현황.csv', encoding='utf-8-sig')

# 열 이름 확인 및 수정
print(df_shelter.columns)  # 열 이름 확인
df_shelter.columns = df_shelter.columns.str.strip()  # 공백 제거

# '대피소 종류'와 '내진설계적용여부'별 데이터 개수 집계
summary = df_shelter.groupby(['대피소 종류', '내진설계적용여부']).size().reset_index(name='개수')

# 세로 막대 그래프 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=summary, x='대피소 종류', y='개수', hue='내진설계적용여부', palette=['salmon', 'skyblue'])

# 그래프 꾸미기
plt.title('대피소 종류별 내진설계 적용 현황', fontsize=16)
plt.xlabel('대피소 종류', fontsize=14)
plt.ylabel('대피소 개수', fontsize=14)
plt.legend(title='내진설계적용여부', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 그래프 출력
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 도로명전체주소에서 광역시/도 정보 추출
df_shelter_min['광역시도'] = df_shelter_min['도로명전체주소'].str.extract(
    r'(서울특별시|부산광역시|대구광역시|인천광역시|광주광역시|대전광역시|울산광역시|세종특별자치시|경기도|강원특별자치도|충청북도|충청남도|전북특별자치도|전라남도|경상북도|경상남도|제주특별자치도)'
)

# 광역시도별 대피소 수 계산
shelter_counts_by_province = df_shelter_min['광역시도'].value_counts()

# 그래프 설정
plt.figure(figsize=(12, 8))
sns.barplot(x=shelter_counts_by_province.values, y=shelter_counts_by_province.index, orient='h', palette='coolwarm')

# 그래프 꾸미기
plt.title('광역시도별 대피소 수', fontsize=16)
plt.xlabel('대피소 수', fontsize=12)
plt.ylabel('광역시도', fontsize=12)

# 각 막대에 숫자 표시
for i, v in enumerate(shelter_counts_by_province.values):
    plt.text(v + 0.5, i, str(v), va='center')

plt.tight_layout()
plt.show()

# 광역시도별 대피소 수 출력
print("광역시도별 대피소 수:")
print(shelter_counts_by_province)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 파일 읽기
df_nuclear_50 = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/원자력발전소/국내원자력발전소_50km이내_지진발생현황.csv', encoding='EUC-KR')

# 데이터 확인
df_nuclear_50['연도'] = pd.to_datetime(df_nuclear_50['발생년월일']).dt.year  # Assuming '발생일자' contains the date

# 거리 구간 설정
bins = [0, 10, 20, 30, 40, 50]
labels = ['0-10km', '10-20km', '20-30km', '30-40km', '40-50km']
df_nuclear_50['거리 구간'] = pd.cut(df_nuclear_50['거리(km)'], bins=bins, labels=labels)

# 연도 및 거리 구간별 지진 규모 평균 계산
distance_magnitude_year = df_nuclear_50.groupby(['연도', '거리 구간'])['규모'].mean().reset_index()

# 시각화
plt.figure(figsize=(14, 8))
sns.lineplot(data=distance_magnitude_year, x='연도', y='규모', hue='거리 구간', marker='o', linewidth=2)

# 그래프 꾸미기
plt.title('연도별 원전과의 거리 구간별 평균 지진 규모', fontsize=16)
plt.xlabel('연도', fontsize=14)
plt.ylabel('평균 지진 규모', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='거리 구간', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```
