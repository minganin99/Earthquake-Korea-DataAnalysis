##### 아래 코드는 규모가 3이상인 지진이 발생한 곳의 위치 리스트입니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import folium
from folium.plugins import MarkerCluster
from IPython.display import display

# 데이터 로드 (여기서는 'r' 데이터프레임이 이미 있다고 가정합니다)
# r = pd.read_csv('your_earthquake_data.csv')

# 위도와 경도 데이터 전처리 함수
def parse_coordinates(coord_str):
    parts = coord_str.split()
    if len(parts) == 2:
        value, direction = parts
        value = float(value)
        if direction in ['S', 'W']:
            value = -value
        return value
    return None

# 위도와 경도 데이터 전처리
r['위도_숫자'] = r['위도'].apply(parse_coordinates)
r['경도_숫자'] = r['경도'].apply(parse_coordinates)

# 이상치 계산
Q1 = r['규모'].quantile(0.25)
Q3 = r['규모'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = r[(r['규모'] < lower_bound) | (r['규모'] > upper_bound)]

print("\n이상치 데이터:")
if len(outliers) > 0:
    for index, row in outliers.iterrows():
        print(f"발생시각: {row['발생시각']}, 규모: {row['규모']:.2f}, 위치: {row['위치']}")
else:
    print("이상치가 없습니다.")

print(f"\n이상치 개수: {len(outliers)}")

# 이상치를 제외한 데이터의 통계 정보
normal_data = r[(r['규모'] >= lower_bound) & (r['규모'] <= upper_bound)]
print("\n이상치를 제외한 데이터 통계:")
print(f"데이터 개수: {len(normal_data)}")
print(f"규모 평균: {normal_data['규모'].mean():.2f}")
print(f"규모 중앙값: {normal_data['규모'].median():.2f}")
print(f"규모 최대값: {normal_data['규모'].max():.2f}")

# 규모가 큰 상위 20개 지진 선택
top_20_earthquakes = r.sort_values('규모', ascending=False).head(20)

# 한국의 중심 좌표 (대략적인 위치)
korea_center = [36.5, 127.5]

# 지도 생성
m = folium.Map(location=korea_center, zoom_start=7)

# 마커 클러스터 생성
marker_cluster = MarkerCluster().add_to(m)

# 각 지진에 대한 마커 추가
for idx, row in top_20_earthquakes.iterrows():
    if pd.notnull(row['위도_숫자']) and pd.notnull(row['경도_숫자']):
        folium.Marker(
            location=[row['위도_숫자'], row['경도_숫자']],
            popup=f"발생시각: {row['발생시각']}<br>규모: {row['규모']}<br>깊이: {row['깊이']}km<br>진도: {row['진도']}<br>위치: {row['위치']}",
            tooltip=f"규모 {row['규모']} 지진"
        ).add_to(marker_cluster)

# 지도 표시
display(m)

# 결과 출력
print("\n규모가 큰 상위 20개 지진:")
print(top_20_earthquakes[['발생시각', '규모', '깊이', '진도', '위치', '위도_숫자', '경도_숫자']])

# 기본 통계 정보 출력
print("\n상위 20개 지진의 기본 통계:")
print(top_20_earthquakes['규모'].describe())

# 연도별 상위 20개 지진 분포
year_distribution = top_20_earthquakes['발생시각'].dt.year.value_counts().sort_index()
print("\n연도별 상위 20개 지진 분포:")
print(year_distribution)

# 지역별 상위 20개 지진 분포
location_distribution = top_20_earthquakes['위치'].value_counts()
print("\n지역별 상위 20개 지진 분포:")
print(location_distribution)

# 규모 분포 히스토그램
plt.figure(figsize=(10, 6))
sns.histplot(r['규모'], bins=20, kde=True)
plt.title('지진 규모 분포')
plt.xlabel('규모')
plt.ylabel('빈도')
plt.show()

# 연도별 지진 발생 추이
r['연도'] = pd.to_datetime(r['발생시각']).dt.year
yearly_earthquakes = r.groupby('연도').size()
plt.figure(figsize=(12, 6))
yearly_earthquakes.plot(kind='line')
plt.title('연도별 지진 발생 추이')
plt.xlabel('연도')
plt.ylabel('지진 발생 횟수')
plt.show()
```

#### 이를 추세선으로 표시해보면

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지


# '발생시각' 열을 datetime 형식으로 변환
df['발생시각'] = pd.to_datetime(df['발생시각'])

# 규모 3 이상의 지진 데이터만 필터링
df_filtered = df[df['규모'] >= 3.0]

# 결측치 제거
df_filtered = df_filtered.dropna(subset=['발생시각', '규모'])

# 데이터를 시간순으로 정렬
df_filtered = df_filtered.sort_values('발생시각')

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 산점도 그리기
plt.scatter(df_filtered['발생시각'], df_filtered['규모'], alpha=0.6)

# 추세선 추가
x = df_filtered['발생시각'].astype('int64') // 10**9  # datetime을 Unix timestamp로 변환
y = df_filtered['규모']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept
plt.plot(df_filtered['발생시각'], line, color='red', label=f'추세선 (R² = {r_value**2:.3f})')

# 그래프 꾸미기
plt.title('시간에 따른 규모 3 이상 지진의 규모 변화')
plt.xlabel('발생 시각')
plt.ylabel('지진 규모')
plt.legend()

# y축 범위 설정 (규모 3부터 시작)
plt.ylim(bottom=3)

# x축 날짜 포맷 설정
plt.gcf().autofmt_xdate()

# 그래프 표시
plt.show()

# 추가적인 통계 정보 출력
print(f"기울기: {slope:.6f}")
print(f"y 절편: {intercept:.6f}")
print(f"R-squared: {r_value**2:.3f}")
print(f"p-value: {p_value:.6f}")

# 데이터 요약 정보
print("\n데이터 요약:")
print(f"규모 3 이상 지진 총 발생 횟수: {len(df_filtered)}")
print(f"첫 번째 지진 발생 시각: {df_filtered['발생시각'].min()}")
print(f"마지막 지진 발생 시각: {df_filtered['발생시각'].max()}")
print(f"평균 규모: {df_filtered['규모'].mean():.2f}")
print(f"최대 규모: {df_filtered['규모'].max():.2f}")
print(f"최소 규모: {df_filtered['규모'].min():.2f}")
```

- 오클라호마 지진과 포항 지진의 공통점

오클라호마 지진과 포항 지진의 주요 공통점은 다음과 같습니다:

1. 인공적 요인에 의한 유발지진:

두 지진 모두 인간의 활동으로 인해 발생한 유발지진으로 추정됩니다[1][2]. 오클라호마의 경우 석유 산업의 폐수 주입이, 포항의 경우 지열발전을 위한 물 주입이 원인으로 지목되었습니다.

2. 유체 주입과의 연관성:

두 지역 모두 지하에 유체(폐수 또는 물)를 주입하는 과정이 지진 발생과 밀접한 관련이 있는 것으로 나타났습니다[1][4]. 유체 주입이 지반의 압력과 응력 변화를 일으켜 지진을 촉발했다는 분석이 있습니다.

3. 지진 발생 빈도 증가:

두 지역 모두 유체 주입 활동 이후 지진 발생 빈도가 급격히 증가했습니다[1][4]. 오클라호마의 경우 규모 3급 지진이 연간 1-2회에서 일일 1-2회로 증가했고, 포항의 경우 이전에 거의 없던 지진이 빈번하게 발생하기 시작했습니다.

4. 과학적 논란과 연구:

두 지진 모두 초기에는 원인에 대한 논란이 있었으나, 이후 과학적 연구를 통해 인공적 요인이 주요 원인일 가능성이 높다는 결론에 도달했습니다[1][4][5].

5. 정책적 대응:

두 사례 모두 지진 발생 후 관련 활동(폐수 주입 또는 지열발전)에 대한 규제나 중단 조치가 이루어졌습니다[1][5].

이러한 공통점들은 인간의 지하 활동이 지진 발생에 미치는 영향에 대한 중요한 사례로, 향후 유사한 프로젝트를 계획할 때 신중한 접근이 필요함을 시사합니다.

```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster
from branca.colormap import LinearColormap

# 데이터 로드 및 전처리
df_japan = pd.read_csv('시계열_일본_지진데이터.csv')

def parse_coordinates(coord_str):
    if pd.isna(coord_str):
        return None
    parts = str(coord_str).split()
    if len(parts) == 2:
        value, direction = parts
        try:
            value = float(value)
            if direction in ['S', 'W']:
                value = -value
            return value
        except ValueError:
            return None
    return None

df_japan['위도_숫자'] = df_japan['위도'].apply(parse_coordinates)
df_japan['경도_숫자'] = df_japan['경도'].apply(parse_coordinates)
df_japan['규모'] = pd.to_numeric(df_japan['규모'], errors='coerce')
df_japan['깊이(km)'] = pd.to_numeric(df_japan['깊이(km)'], errors='coerce')

# 유효한 데이터만 선택 (위도, 경도, 규모가 모두 있는 경우)
valid_earthquakes = df_japan.dropna(subset=['위도_숫자', '경도_숫자', '규모'])

# 지도 생성
m = folium.Map(location=[36, 138], zoom_start=5)

# 색상 맵 생성
colormap = LinearColormap(colors=['blue', 'green', 'yellow', 'orange', 'red'], 
                          vmin=valid_earthquakes['규모'].min(), 
                          vmax=valid_earthquakes['규모'].max())

# 마커 클러스터 생성
marker_cluster = MarkerCluster().add_to(m)

# 각 지진에 대한 마커 추가
for idx, row in valid_earthquakes.iterrows():
    folium.CircleMarker(
        location=[row['위도_숫자'], row['경도_숫자']],
        radius=row['규모'] * 1.5,  # 규모에 따라 원의 크기 조절
        popup=f"발생시각: {row['발생시각']}<br>규모: {row['규모']:.1f}<br>깊이: {row['깊이(km)']}km<br>위치: {row['위치']}",
        tooltip=f"규모 {row['규모']:.1f} 지진",
        color=colormap(row['규모']),
        fill=True,
        fillColor=colormap(row['규모']),
        fillOpacity=0.7
    ).add_to(marker_cluster)

# 색상 범례 추가
colormap.add_to(m)

# 한국의 대략적인 위치에 마커 추가
folium.Marker(
    location=[37.5665, 126.9780],
    popup="Seoul, South Korea",
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(m)

# 지도 표시
display(m)

# 기본 통계 정보 출력
print("\n지진의 기본 통계:")
print(valid_earthquakes['규모'].describe())

# 연도별 지진 분포
valid_earthquakes['발생시각'] = pd.to_datetime(valid_earthquakes['발생시각'])
year_distribution = valid_earthquakes['발생시각'].dt.year.value_counts().sort_index()
print("\n연도별 지진 분포:")
print(year_distribution)

# 규모별 지진 분포
magnitude_distribution = pd.cut(valid_earthquakes['규모'], bins=[0, 3, 4, 5, 6, 7, np.inf], labels=['0-3', '3-4', '4-5', '5-6', '6-7', '7+'])
print("\n규모별 지진 분포:")
print(magnitude_distribution.value_counts().sort_index())
```

```python
import pandas as pd
from datetime import timedelta

# 일본 지진 데이터
japan_earthquakes = pd.DataFrame({
    '발생시각': ['2019-06-18 22:22:00', '2022-01-22 01:08:00', '2023-05-05 14:42:00', 
                 '2024-04-17 23:14:00', '2024-01-09 17:59:00'],
    '규모': [6.8, 6.4, 6.3, 6.4, 6.0]
})
japan_earthquakes['발생시각'] = pd.to_datetime(japan_earthquakes['발생시각'])

# 한국 지진 데이터 로드
df_korea = pd.read_csv('C:/Users/campus4D036/Desktop/프로젝트/팀프로젝트/Earthquake-Korea-DataAnalysis/찐최종 데이터셋/지진/시계열_한국_지진데이터_찐최종.csv')
df_korea['발생시각'] = pd.to_datetime(df_korea['발생시각'])

# 규모와 깊이를 숫자로 변환
df_korea['규모'] = pd.to_numeric(df_korea['규모'], errors='coerce')
df_korea['깊이(km)'] = pd.to_numeric(df_korea['깊이'], errors='coerce')

# 일본 지진 발생 후 7일 이내의 한국 지진 찾기
def find_related_earthquakes(japan_quake, korea_df, days=7):
    start_time = japan_quake['발생시각']
    end_time = start_time + timedelta(days=days)
    related_quakes = korea_df[(korea_df['발생시각'] > start_time) & (df_korea['발생시각'] <= end_time)]
    return related_quakes

# 각 일본 지진에 대해 관련된 한국 지진 찾기
for _, japan_quake in japan_earthquakes.iterrows():
    print(f"\n일본 지진 발생: {japan_quake['발생시각']} (규모 {japan_quake['규모']})")
    related_quakes = find_related_earthquakes(japan_quake, df_korea)
    
    if len(related_quakes) > 0:
        print("관련된 한국 지진:")
        for _, korea_quake in related_quakes.iterrows():
            time_diff = korea_quake['발생시각'] - japan_quake['발생시각']
            try:
                print(f"  - {korea_quake['발생시각']} (규모 {korea_quake['규모']:.1f}, "
                      f"깊이 {korea_quake['깊이(km)']:.1f}km, 위치: {korea_quake['위치']}, "
                      f"일본 지진 후 {time_diff.total_seconds() / 3600:.1f}시간 후)")
            except ValueError:
                print(f"  - {korea_quake['발생시각']} (규모 {korea_quake['규모']}, "
                      f"깊이 {korea_quake['깊이(km)']}km, 위치: {korea_quake['위치']}, "
                      f"일본 지진 후 {time_diff.total_seconds() / 3600:.1f}시간 후)")
    else:
        print("7일 이내에 관련된 한국 지진 없음")

# 통계 정보 출력
total_related_quakes = sum(len(find_related_earthquakes(quake, df_korea)) for _, quake in japan_earthquakes.iterrows())
print(f"\n총 관련 한국 지진 수: {total_related_quakes}")
print(f"분석된 일본 지진 수: {len(japan_earthquakes)}")
```

##### 전국 공공 건축물 내진설계 현황

```python
df_pu = pd.read_csv('찐최종 데이터셋\대책현황\내진설계\국내공공건축물_내진설계현황_EDA.csv', )
df_pu
```

```python
# 내진 설계 적용 비율 계산
df_pu['내진율(동수)'] = (df_pu['내진설계 적용(동수)'] / df_pu['전체 공공건축물(동수)']) * 100
df_pu['내진율(연면적)'] = (df_pu['내진설계 적용(연면적)'] / df_pu['전체 공공건축물(연면적)']) * 100

# 그래프 설정
plt.figure(figsize=(14, 7))

# 동수 기준 내진율 시각화
sns.barplot(x='지역', y='내진율(동수)', data=df_pu, color='blue', alpha=0.6, label='내진율(동수)')

# 연면적 기준 내진율 시각화
sns.barplot(x='지역', y='내진율(연면적)', data=df_pu, color='orange', alpha=0.6, label='내진율(연면적)')

# 그래프 제목 및 레이블 설정
plt.title('지역별 내진 설계 적용 비율 비교')
plt.xlabel('지역')
plt.ylabel('내진율 (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y')
plt.show()
```

##### 연도별 내진 설계율

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 준공일을 안전하게 datetime 형식으로 변환하는 함수
def safe_date_parse(date_val):
    if pd.isna(date_val):
        return None
    try:
        date_str = str(int(date_val))
        if len(date_str) == 8:  # YYYYMMDD 형식
            return datetime.strptime(date_str, '%Y%m%d')
        elif len(date_str) == 6:  # YYYYMM 형식
            return datetime.strptime(date_str, '%Y%m')
        else:
            return None
    except ValueError:
        return None

# 준공일 변환 적용
df_resist_home['준공일'] = df_resist_home['준공일'].apply(safe_date_parse)

# 유효하지 않은 날짜 제거
df_resist_home = df_resist_home.dropna(subset=['준공일'])

# 연도별 구간 정의 함수
def get_year_group(date):
    year = date.year
    if year < 1980:
        return '1980년대 이전'
    elif year < 1990:
        return '1980년대'
    elif year < 2000:
        return '1990년대'
    elif year < 2010:
        return '2000년대'
    else:
        return '2010년 이후'

# 연도 구간 열 추가
df_resist_home['연도구간'] = df_resist_home['준공일'].apply(get_year_group)

# 연도구간별 내진설계 대상여부 비율과 수 계산
year_group_counts = df_resist_home.groupby(['연도구간', '내진설계 대상여부']).size().unstack(fill_value=0)
year_group_ratios = year_group_counts.div(year_group_counts.sum(axis=1), axis=0)

# 결과 출력
print("연도구간별 내진설계 대상여부 비율:")
print(year_group_ratios)
print("\n연도구간별 내진설계 대상여부 수:")
print(year_group_counts)

# 파이 그래프로 시각화
fig, axes = plt.subplots(2, 3, figsize=(20, 15))
axes = axes.flatten()

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val:,d})'
    return my_autopct

for i, (year_group, data) in enumerate(year_group_counts.iterrows()):
    if i >= len(axes):
        break
    ax = axes[i]
    wedges, texts, autotexts = ax.pie(data, autopct=make_autopct(data), startangle=90)
    
    # 범례 추가
    ax.legend(wedges, data.index, title="내진설계 대상여부", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    # 그래프 아래에 정보 표시
    total = data.sum()
    info_text = f"{year_group}\n총 건물 수: {total:,d}"
    ax.text(0, -1.3, info_text, ha='center', va='center', fontsize=10)

# 사용하지 않는 서브플롯 제거
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('연도구간별 내진설계 대상여부 비율 및 수', fontsize=16)
plt.tight_layout()
plt.show()

# 각 연도구간별 총 데이터 수 계산
total_counts = year_group_counts.sum(axis=1)
print("\n연도구간별 총 데이터 수:")
print(total_counts)
```

### 내진 설계에 대한 관심도가 늘어나는 모습을 알 수 있습니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 지역 분류 함수
def classify_region(address):
    regions = ['서울특별시', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도', '대구광역시', '부산광역시', '인천광역시', '광주광역시', '대전광역시']
    for region in regions:
        if region in str(address):
            return region
    return '기타'

# 지역 분류 적용
df_resist_home['지역'] = df_resist_home['대지위치'].apply(classify_region)

# 지정된 도시들만 선택
df_resist_cities = df_resist_home[df_resist_home['지역'] != '기타']

# 지역별 내진설계 대상여부 수 계산
region_counts = df_resist_cities.groupby(['지역', '내진설계 대상여부']).size().unstack(fill_value=0)

# 주택 수 기준으로 정렬
region_counts['총 주택수'] = region_counts.sum(axis=1)
region_counts_sorted = region_counts.sort_values('총 주택수', ascending=False)

# 파이 그래프 생성 함수
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val:,d})'
    return my_autopct

# 파이 그래프로 시각화
fig, axes = plt.subplots(4, 3, figsize=(20, 24))
axes = axes.flatten()

for i, (region, data) in enumerate(region_counts_sorted.iterrows()):
    if i >= len(axes):
        break
    ax = axes[i]
    wedges, texts, autotexts = ax.pie(data.drop('총 주택수'), 
                                      autopct=make_autopct(data.drop('총 주택수')), 
                                      startangle=90)
    
    # 지역명을 파이 그래프 아래에 표시
    ax.text(0, -1.2, f'{region}\n(총 {data["총 주택수"]:,d})', ha='center', va='center', fontsize=10)
    
    # 범례 추가
    ax.legend(wedges, data.drop('총 주택수').index, 
              title="내진설계 대상여부", 
              loc="center left", 
              bbox_to_anchor=(1, 0, 0.5, 1))

# 사용하지 않는 서브플롯 제거
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('지역별 내진설계 대상여부 비율 및 수', fontsize=16)
plt.tight_layout()
plt.show()

# 각 지역별 총 데이터 수 출력
print("\n지역별 총 주택 수:")
print(region_counts_sorted['총 주택수'].sort_values(ascending=False))
```

#### 지역별 내진 설계율에 대해서 알아봤습니다
여기서 지진이 자주 발생하는 지역에 대한 내진 설계율을 알아보면 좋을 거 같아요

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지
# 연면적 정규 분포 파라미터 설정
mean_area = 1000  # 평균 연면적 (단위: m²)
std_dev_area = 200  # 연면적의 표준편차

# 정규 분포를 따르는 연면적 데이터 생성 (예: 1000개의 샘플)
num_samples = 1000
area_distribution = np.random.normal(mean_area, std_dev_area, num_samples)

# 음수 값 제거 (연면적은 항상 양수이어야 함)
area_distribution = np.maximum(area_distribution, 0)

# DataFrame 생성
area_df = pd.DataFrame(area_distribution, columns=['연면적'])

# 그래프 설정
plt.figure(figsize=(12, 6))
sns.histplot(area_df['연면적'], bins=30, kde=True, color='skyblue')

# 그래프 꾸미기
plt.title('연면적의 정규 분포', fontsize=16)
plt.xlabel('연면적 (m²)', fontsize=12)
plt.ylabel('빈도', fontsize=12)
plt.axvline(x=mean_area, color='red', linestyle='--', label='평균')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# 평균값 표시
plt.text(mean_area, plt.gca().get_ylim()[1], f'평균: {mean_area:.2f} m²', 
         horizontalalignment='center', verticalalignment='bottom')

plt.show()

# 기술 통계량 출력
print(area_df['연면적'].describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 연면적 구간 설정 (0~400, 400~800, 800~1200, 1200~1600, 1600 이상)
bins = [0, 400, 800, 1200, 1600, float('inf')]
labels = ['0~400㎡', '400~800㎡', '800~1200㎡', '1200~1600㎡', '1600㎡ 이상']

# 연면적 구간별로 데이터 분류
df_resist_home['연면적 구간'] = pd.cut(df_resist_home['연면적'], bins=bins, labels=labels)

# 각 구간별 내진설계 대상여부 개수 계산
grouped = df_resist_home.groupby(['연면적 구간', '내진설계 대상여부']).size().unstack(fill_value=0)

# 비율 계산
grouped_ratio = grouped.div(grouped.sum(axis=1), axis=0)

# 결과 출력
print("연면적 구간별 내진설계 대상여부:")
print(grouped)
print("\n연면적 구간별 내진설계 대상여부 비율:")
print(grouped_ratio)

# 파이 차트로 시각화
fig, axes = plt.subplots(2, 3, figsize=(20, 15))
axes = axes.flatten()

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val:,d})'
    return my_autopct

for i, (group_name, group_data) in enumerate(grouped.iterrows()):
    if i >= len(axes):
        break
    ax = axes[i]
    wedges, texts, autotexts = ax.pie(group_data, labels=group_data.index, 
                                      autopct=make_autopct(group_data), 
                                      startangle=90)
    
    # 범례 추가
    ax.legend(wedges, group_data.index, 
              title="내진설계 대상여부", 
              loc="center left", 
              bbox_to_anchor=(1, 0, 0.5, 1))
    
    # 그래프 아래에 정보 표시
    total = group_data.sum()
    info_text = f"{group_name}\n총 건물 수: {total:,d}"
    ax.text(0, -1.3, info_text, ha='center', va='center', fontsize=10)

# 사용하지 않는 서브플롯 제거
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('연면적 구간별 내진설계 대상여부 비율 및 수', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import platform

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 500 제곱미터 이하의 건물만 선택
df_small = df_resist_home[df_resist_home['연면적'] <= 500]

# 연면적 구간 정의 함수
def get_area_group(area):
    if area <= 100:
        return '0-100㎡'
    elif area <= 200:
        return '101-200㎡'
    elif area <= 300:
        return '201-300㎡'
    elif area <= 400:
        return '301-400㎡'
    else:
        return '401-500㎡'

# 연면적 구간 열 추가
df_small['연면적구간'] = df_small['연면적'].apply(get_area_group)

# 연면적구간별 내진설계 대상여부 개수 계산
grouped = df_small.groupby(['연면적구간', '내진설계 대상여부']).size().unstack(fill_value=0)

# 비율 계산
grouped_ratio = grouped.div(grouped.sum(axis=1), axis=0)

# 평균 지상 층수 계산
avg_floors = df_small.groupby('연면적구간')['지상층수'].mean()

# 파이 차트로 시각화
fig, axes = plt.subplots(2, 3, figsize=(20, 15))
axes = axes.flatten()

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val:,d})'
    return my_autopct

for i, (group_name, group_data) in enumerate(grouped.iterrows()):
    if i >= len(axes):
        break
    ax = axes[i]
    wedges, texts, autotexts = ax.pie(group_data, autopct=make_autopct(group_data), startangle=90)
    
    # 범례 추가
    ax.legend(wedges, group_data.index, title="내진설계 대상여부", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    # 그래프 아래에 정보 표시
    total = group_data.sum()
    info_text = (f"{group_name}\n"
                 f"총 건물 수: {total:,d}\n"
                 f"평균 지상층수: {avg_floors[group_name]:.1f}")
    ax.text(0, -1.5, info_text, ha='center', va='center', fontsize=10)

# 사용하지 않는 서브플롯 제거
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('연면적구간별 내진설계 대상여부 비율 및 수 (500㎡ 이하)', fontsize=16)
plt.tight_layout()
plt.show()

# 결과 출력
print("연면적구간별 내진설계 대상여부 수:")
print(grouped)
print("\n연면적구간별 내진설계 대상여부 비율:")
print(grouped_ratio)
print("\n연면적구간별 평균 지상층수:")
print(avg_floors)
```
