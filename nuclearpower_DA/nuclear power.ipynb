```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df_excel = pd.read_excel('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/대피소/대피소(해안)_찐최종.xlsx')
df_excel = pd.read_excel('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/대피소/국내_민방위대피소현황.xlsx')

# 필요한 컬럼 추출 및 데이터 통합
df_csv_filtered = df_csv[['대피장소명']].copy()
df_excel_filtered = df_excel[['시설명', '시설면적(㎡)']].copy()

# 병합을 위해 컬럼명 통일
df_excel_filtered = df_excel_filtered.rename(columns={'시설명': '대피장소명'})
df = pd.merge(df_csv_filtered, df_excel_filtered, on='대피장소명', how='inner')

# 결측치 제거 후 확인
df = df.dropna(subset=['시설면적(㎡)', '대피장소명'])

# 대피소수 및 면적당 대피소 수 계산
df['대피소수'] = df.groupby('대피장소명')['대피장소명'].transform('count')
df['면적당_대피소수'] = df['대피소수'] / df['시설면적(㎡)']
df_unique = df[['대피장소명', '면적당_대피소수']].drop_duplicates()

# 시각화: 대피장소명별 면적당 대피소 수
plt.figure(figsize=(12, 8))

# 상위 10개 대피소만 선택하여 시각화 (예시)
top_10_df = df_unique.nlargest(10, '면적당_대피소수')

sns.barplot(data=top_10_df, x='대피장소명', y='면적당_대피소수', color="skyblue")

# X축 라벨 회전 및 레이아웃 조정
plt.xticks(rotation=90)
plt.title('시설 면적당 대피소 수 (상위 10개 대피소)')
plt.xlabel('대피장소명')
plt.ylabel('면적당 대피소 수 (대피소 수/㎡)')
plt.tight_layout()  # 레이아웃 조정

# 시각화 출력
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
f_earthquake = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/지진/시계열_한국_지진데이터_찐최종.csv')
df_shelters = pd.read_excel('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/대피소/대피소(해안)_찐최종.xlsx')

# 지진 발생 빈도수 계산 (위치별로)
earthquake_freq = f_earthquake['위치'].value_counts().reset_index()
earthquake_freq.columns = ['위치', '지진빈도수']

# 대피소 수 계산 (위치별로)
shelter_count = df_shelters['대피장소상세위치'].value_counts().reset_index()
shelter_count.columns = ['위치', '대피소수']

# 두 데이터 병합 (위치 기준)
df_merged = pd.merge(earthquake_freq, shelter_count, on='위치', how='inner')

# 결측값 제거 (만약 있으면)
df_merged = df_merged.dropna(subset=['지진빈도수', '대피소수'])

# 시각화: 산점도
plt.figure(figsize=(14, 8))
sns.scatterplot(x='지진빈도수', y='대피소수', data=df_merged, hue='위치', palette='viridis', s=100, edgecolor='w')

# 그래프 제목 및 레이블
plt.title('지진 발생 빈도수와 대피소 수의 관계', fontsize=16)
plt.xlabel('지진 빈도수', fontsize=14)
plt.ylabel('대피소 수', fontsize=14)
plt.legend(title='위치', bbox_to_anchor=(1, 1))

# 그래프 출력
plt.tight_layout()

# 그래프가 제대로 출력되도록 'plt.show()' 호출
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager

# 한글 폰트 설정 (Windows 환경에서 사용하는 예시)
font_path = "C:/Windows/Fonts/malgun.ttf"  # '맑은 고딕' 폰트 경로
font_prop = font_manager.FontProperties(fname=font_path)

# 데이터 로드
df = pd.read_excel('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/대피소/국내_민방위대피소현황.xlsx')

# 서울, 인천, 경기도를 수도권으로 구분할 키워드
keywords = ['서울', '인천', '경기']

# 지역 구분 함수
def categorize_region(address):
    if '서울' in address or '인천' in address or '경기' in address:
        return '수도권'
    else:
        return '기타지역'

# '소재지전체주소'에서 지역 추출하여 새로운 '지역' 컬럼 생성
df['지역'] = df['소재지전체주소'].apply(categorize_region)

# 지역별 시설 개수 파악
region_counts = df['지역'].value_counts()

# 결과 시각화 (막대 그래프)
plt.figure(figsize=(8, 5))
region_counts.plot(kind='bar', color='lightblue')

# 그래프 제목 및 레이블 설정 (한글 폰트 적용)
plt.title('지역별 시설 개수', fontsize=16, fontproperties=font_prop)
plt.xlabel('지역', fontsize=14, fontproperties=font_prop)
plt.ylabel('시설 개수', fontsize=14, fontproperties=font_prop)
plt.xticks(rotation=45, fontsize=12, fontproperties=font_prop)
plt.grid(True)

# 그래프 표시
plt.show()

# 지역별 개수 출력
print(region_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
df = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/원자력발전소/한국원자력발전소_지진감시이력_EDA.csv')

# 데이터 전처리: 연도별 지진동 횟수 데이터를 변환
df_melted = df.melt(id_vars=['구분', '지역'], var_name='연도', value_name='지진동 횟수')
df_melted['연도'] = df_melted['연도'].astype(int)  # 연도를 정수형으로 변환

# 시각화
plt.figure(figsize=(14, 8))

# 발전소별로 다른 선 그래프 생성
for plant in df_melted['지역'].unique():
    plant_data = df_melted[df_melted['지역'] == plant]
    plt.plot(plant_data['연도'], plant_data['지진동 횟수'], marker='o', label=plant)

# 그래프 세부 설정
plt.title('발전소별 연도별 지진동 횟수 추이', fontsize=16)
plt.xlabel('연도', fontsize=14)
plt.ylabel('지진동 횟수', fontsize=14)
plt.legend(title='발전소', title_fontsize=12, fontsize=10)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)

# 그래프 표시
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
df = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/원자력발전소/국내원자력발전소_내진설계현황_EDA.csv')

# '내진설계기준'과 '내진성능향상값'을 숫자형으로 변환
df['내진설계기준'] = pd.to_numeric(df['내진설계기준'], errors='coerce')
df['내진성능향상값'] = pd.to_numeric(df['내진성능향상값'], errors='coerce')

# NaN 값 제거
df_clean = df.dropna(subset=['호기', '내진설계기준', '내진성능향상값'])

# 호기별 내진설계기준과 내진성능향상값 분석
analysis = df_clean.groupby('호기')[['내진설계기준', '내진성능향상값']].mean().reset_index()

# 시각화 (세로 막대 그래프)
plt.figure(figsize=(14, 8))

# 두 변수를 하나의 그래프에 표현
analysis_melted = analysis.melt(id_vars='호기', value_vars=['내진설계기준', '내진성능향상값'], 
                                var_name='항목', value_name='값')

# 세로 막대 그래프
sns.barplot(data=analysis_melted, x='호기', y='값', hue='항목', palette='coolwarm')

# 제목 및 레이블 설정
plt.title('호기별 내진설계기준과 내진성능향상값', fontsize=18, fontweight='bold')
plt.xlabel('호기', fontsize=14)
plt.ylabel('값', fontsize=14)

# 범례 설정
plt.legend(title='항목', fontsize=12)

# X, Y 축 레이블 설정
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# 레이아웃 최적화
plt.tight_layout()

# 그래프 표시
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
df = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/원자력발전소/국내원자력발전소_50km이내_지진발생현황.csv', encoding='EUC-KR')

# 발생 깊이와 원전 거리 간의 관계 시각화 (산점도)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='거리(km)', y='깊이(km)', color='b', marker='o')
plt.title('발생 깊이와 원전 거리 간의 관계', fontsize=16)
plt.xlabel('원전과의 거리 (km)', fontsize=14)
plt.ylabel('발생 깊이 (km)', fontsize=14)
plt.grid(True)

# 상관계수 계산
correlation = df['거리(km)'].corr(df['깊이(km)'])

# 상관계수 오른쪽 상단에 표시
plt.text(0.95, 0.95, f'상관계수: {correlation:.2f}', ha='right', va='top', transform=plt.gca().transAxes, fontsize=12, color='red')

# 그래프 표시
plt.show()

# 상관계수 출력
print(f'발생 깊이와 원전 거리 간의 상관계수: {correlation:.2f}')


```

### 원전과 가까운 곳에서 발생한 지진은 깊이가 얕은지 확인합니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
df = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/원자력발전소/국내원자력발전소_50km이내_지진발생현황.csv', encoding='EUC-KR')

# 거리 구간 설정 (예: 0-10, 10-20, ..., 40-50 km)
bins = [0, 10, 20, 30, 40, 50]
labels = ['0-10km', '10-20km', '20-30km', '30-40km', '40-50km']
df['거리 구간'] = pd.cut(df['거리(km)'], bins=bins, labels=labels)

# 거리 구간별 지진 규모 평균 계산
distance_magnitude = df.groupby('거리 구간')['규모'].mean().reset_index()

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=distance_magnitude, x='거리 구간', y='규모', palette='viridis')
plt.title('원전과의 거리 구간별 평균 지진 규모', fontsize=16)
plt.xlabel('거리 구간', fontsize=14)
plt.ylabel('평균 지진 규모', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 원전과의 거리 구간에 따라 지진 규모

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
df = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/원자력발전소/국내원자력발전소_50km이내_지진발생현황.csv', encoding='EUC-KR')

# 날짜에서 연도 추출
df['발생년'] = pd.to_datetime(df['발생년월일'], errors='coerce').dt.year

# 연도별 지진 발생 횟수 집계
yearly_earthquake_counts = df.groupby('발생년').size()

# 연도별 원전 인근 가속도의 평균값 계산
yearly_avg_acceleration = df.groupby('발생년')['원전인근가속도값(g)'].mean()

# 이중축 그래프 생성
fig, ax1 = plt.subplots(figsize=(10, 6))

# 첫 번째 축: 연도별 지진 발생 횟수
ax1.bar(yearly_earthquake_counts.index, yearly_earthquake_counts.values, color='skyblue', alpha=0.7, label='지진 발생 횟수')
ax1.set_xlabel('연도', fontsize=14)
ax1.set_ylabel('지진 발생 횟수', color='skyblue', fontsize=14)
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.legend(loc='upper left')

# 두 번째 축: 연도별 평균 원전 인근 가속도 값
ax2 = ax1.twinx()
ax2.plot(yearly_avg_acceleration.index, yearly_avg_acceleration.values, color='salmon', marker='o', label='원전 인근 평균 가속도 값')
ax2.set_ylabel('평균 가속도 값 (g)', color='salmon', fontsize=14)
ax2.tick_params(axis='y', labelcolor='salmon')
ax2.legend(loc='upper right')

# 그래프 제목 및 레이아웃 설정
plt.title('연도별 지진 발생 횟수 및 원전 인근 가속도 추이', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```
