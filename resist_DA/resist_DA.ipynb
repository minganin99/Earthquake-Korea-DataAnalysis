##### 연도별 내진 설계율

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 준공일을 안전하게 datetime 형식으로 변환하는 함수
def safe_date_parse(date_val):
    if pd.isna(date_val):
        return None
    try:
        date_str = str(int(date_val))
        if len(date_str) == 8:  # YYYYMMDD 형식
            return datetime.strptime(date_str, '%Y%m%d')
        elif len(date_str) == 6:  # YYYYMM 형식
            return datetime.strptime(date_str, '%Y%m')
        else:
            return None
    except ValueError:
        return None

# 준공일 변환 적용
df_resist_home['준공일'] = df_resist_home['준공일'].apply(safe_date_parse)

# 유효하지 않은 날짜 제거
df_resist_home = df_resist_home.dropna(subset=['준공일'])

# 연도별 구간 정의 함수
def get_year_group(date):
    year = date.year
    if year < 1980:
        return '1980년대 이전'
    elif year < 1990:
        return '1980년대'
    elif year < 2000:
        return '1990년대'
    elif year < 2010:
        return '2000년대'
    else:
        return '2010년 이후'

# 연도 구간 열 추가
df_resist_home['연도구간'] = df_resist_home['준공일'].apply(get_year_group)

# 연도구간별 내진설계 대상여부 비율과 수 계산
year_group_counts = df_resist_home.groupby(['연도구간', '내진설계 대상여부']).size().unstack(fill_value=0)
year_group_ratios = year_group_counts.div(year_group_counts.sum(axis=1), axis=0)

# 결과 출력
print("연도구간별 내진설계 대상여부 비율:")
print(year_group_ratios)
print("\n연도구간별 내진설계 대상여부 수:")
print(year_group_counts)

# 파이 그래프로 시각화
fig, axes = plt.subplots(2, 3, figsize=(20, 15))
axes = axes.flatten()

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val:,d})'
    return my_autopct

for i, (year_group, data) in enumerate(year_group_counts.iterrows()):
    if i >= len(axes):
        break
    ax = axes[i]
    wedges, texts, autotexts = ax.pie(data, autopct=make_autopct(data), startangle=90)
    
    # 범례 추가
    ax.legend(wedges, data.index, title="내진설계 대상여부", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    # 그래프 아래에 정보 표시
    total = data.sum()
    info_text = f"{year_group}\n총 건물 수: {total:,d}"
    ax.text(0, -1.3, info_text, ha='center', va='center', fontsize=10)

# 사용하지 않는 서브플롯 제거
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('연도구간별 내진설계 대상여부 비율 및 수', fontsize=16)
plt.tight_layout()
plt.show()

# 각 연도구간별 총 데이터 수 계산
total_counts = year_group_counts.sum(axis=1)
print("\n연도구간별 총 데이터 수:")
print(total_counts)
```

##### 지역별 내진 설계율

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 지역 분류 함수
def classify_region(address):
    regions = ['서울특별시', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도', '대구광역시', '부산광역시', '인천광역시', '광주광역시', '대전광역시']
    for region in regions:
        if region in str(address):
            return region
    return '기타'

# 지역 분류 적용
df_resist_home['지역'] = df_resist_home['대지위치'].apply(classify_region)

# 지정된 도시들만 선택
df_resist_cities = df_resist_home[df_resist_home['지역'] != '기타']

# 지역별 내진설계 대상여부 수 계산
region_counts = df_resist_cities.groupby(['지역', '내진설계 대상여부']).size().unstack(fill_value=0)

# 주택 수 기준으로 정렬
region_counts['총 주택수'] = region_counts.sum(axis=1)
region_counts_sorted = region_counts.sort_values('총 주택수', ascending=False)

# 파이 그래프 생성 함수
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val:,d})'
    return my_autopct

# 파이 그래프로 시각화
fig, axes = plt.subplots(4, 3, figsize=(20, 24))
axes = axes.flatten()

for i, (region, data) in enumerate(region_counts_sorted.iterrows()):
    if i >= len(axes):
        break
    ax = axes[i]
    wedges, texts, autotexts = ax.pie(data.drop('총 주택수'), 
                                      autopct=make_autopct(data.drop('총 주택수')), 
                                      startangle=90)
    
    # 지역명을 파이 그래프 아래에 표시
    ax.text(0, -1.2, f'{region}\n(총 {data["총 주택수"]:,d})', ha='center', va='center', fontsize=10)
    
    # 범례 추가
    ax.legend(wedges, data.drop('총 주택수').index, 
              title="내진설계 대상여부", 
              loc="center left", 
              bbox_to_anchor=(1, 0, 0.5, 1))

# 사용하지 않는 서브플롯 제거
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('지역별 내진설계 대상여부 비율 및 수', fontsize=16)
plt.tight_layout()
plt.show()

# 각 지역별 총 데이터 수 출력
print("\n지역별 총 주택 수:")
print(region_counts_sorted['총 주택수'].sort_values(ascending=False))
```

##### 500제곱미터 미만 건물의 내진설계 정보

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 준공일을 안전하게 datetime 형식으로 변환하는 함수
def safe_date_parse(date_val):
    if pd.isna(date_val):
        return None
    try:
        date_str = str(int(date_val))
        if len(date_str) == 8:  # YYYYMMDD 형식
            return datetime.strptime(date_str, '%Y%m%d')
        elif len(date_str) == 6:  # YYYYMM 형식
            return datetime.strptime(date_str, '%Y%m')
        else:
            return None
    except ValueError:
        return None

# 준공일 변환 적용
df_resist_home['준공일'] = df_resist_home['준공일'].apply(safe_date_parse)

# 500 제곱미터 이하의 건물만 선택
df_small = df_resist_home[(df_resist_home['연면적'] <= 500) & (df_resist_home['준공일'].notna())]

# 연도별 구간 정의 함수
def get_year_group(date):
    year = date.year
    if year < 1980:
        return '1980년대 이전'
    elif year < 1990:
        return '1980년대'
    elif year < 2000:
        return '1990년대'
    elif year < 2010:
        return '2000년대'
    else:
        return '2010년 이후'

# 연도 구간 열 추가
df_small['연도구간'] = df_small['준공일'].apply(get_year_group)

# 연도구간별 내진설계 대상여부 개수 계산
grouped = df_small.groupby(['연도구간', '내진설계 대상여부']).size().unstack(fill_value=0)

# 비율 계산
grouped_ratio = grouped.div(grouped.sum(axis=1), axis=0)

# 평균 지상 층수 계산
avg_floors = df_small.groupby('연도구간')['지상층수'].mean()

# 파이 차트로 시각화
fig, axes = plt.subplots(2, 3, figsize=(20, 15))
axes = axes.flatten()

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}%\n({val:,d})'
    return my_autopct

for i, (group_name, group_data) in enumerate(grouped.iterrows()):
    if i >= len(axes):
        break
    ax = axes[i]
    wedges, texts, autotexts = ax.pie(group_data, autopct=make_autopct(group_data), startangle=90)
    
    # 범례 추가
    ax.legend(wedges, group_data.index, title="내진설계 대상여부", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    # 그래프 아래에 정보 표시
    total = group_data.sum()
    info_text = (f"{group_name}\n"
                 f"총 건물 수: {total:,d}\n"
                 f"평균 지상층수: {avg_floors[group_name]:.1f}")
    ax.text(0, -1.5, info_text, ha='center', va='center', fontsize=10)

# 사용하지 않는 서브플롯 제거
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('연도구간별 내진설계 대상여부 비율 및 수 (500㎡ 이하)', fontsize=16)
plt.tight_layout()
plt.show()

# 결과 출력
print("연도구간별 내진설계 대상여부 수:")
print(grouped)
print("\n연도구간별 내진설계 대상여부 비율:")
print(grouped_ratio)
print("\n연도구간별 평균 지상층수:")
print(avg_floors)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import platform

# 한글 폰트 설정
if platform.system() == 'Darwin':  # macOS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 데이터 불러오기
df_earthquake = pd.read_excel('C:/Users/campus4D036/Desktop/프로젝트/팀프로젝트/Earthquake-Korea-DataAnalysis/data_set/최종데이터셋/지진/시계열_한국_지진데이터.xlsx')

# 지진 빈도 계산
earthquake_counts = df_earthquake['위치'].value_counts()

# 내진설계 데이터에서 지역 추출 함수
def extract_region(address):
    regions = ['서울특별시', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도', '대구광역시', '부산광역시', '인천광역시', '광주광역시', '대전광역시']
    for region in regions:
        if region in str(address):
            return region
    return '기타'

# 내진설계 데이터에 지역 정보 추가
df_resist_home['지역'] = df_resist_home['대지위치'].apply(extract_region)

# 내진설계 대상여부의 실제 값 확인
print("내진설계 대상여부의 고유 값:", df_resist_home['내진설계 대상여부'].unique())

# 지역별 내진설계 비율 계산 (실제 값에 따라 수정)
region_resist_ratio = df_resist_home.groupby('지역')['내진설계 대상여부'].value_counts(normalize=True).unstack()

# 지진 빈도 상위 10개 지역 선택
top_10_earthquake_regions = earthquake_counts.head(10)

# 결과 데이터프레임 생성
result_df = pd.DataFrame({
    '지진 발생 횟수': top_10_earthquake_regions
})

# 내진설계 대상 비율 열 추가 (실제 값에 따라 수정)
if 'Y' in region_resist_ratio.columns:
    result_df['내진설계 대상 비율'] = region_resist_ratio['Y'].reindex(top_10_earthquake_regions.index).fillna(0)
else:
    print("'Y' 값이 내진설계 대상여부에 없습니다. 데이터를 확인해 주세요.")
    result_df['내진설계 대상 비율'] = 0

# 결과 출력
print("지진 빈도 상위 10개 지역의 내진설계 현황:")
print(result_df)

# 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x=result_df.index, y='지진 발생 횟수', data=result_df, color='skyblue', alpha=0.8)
sns.lineplot(x=result_df.index, y='내진설계 대상 비율', data=result_df, color='red', marker='o')

plt.title('지진 빈도 상위 10개 지역의 지진 발생 횟수와 내진설계 대상 비율', fontsize=16)
plt.xlabel('지역', fontsize=12)
plt.ylabel('지진 발생 횟수 / 내진설계 대상 비율', fontsize=12)
plt.xticks(rotation=45)

# 두 번째 y축 추가
ax2 = plt.twinx()
ax2.set_ylabel('내진설계 대상 비율', color='red', fontsize=12)
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(0, 1)

plt.tight_layout()
plt.show()
```
