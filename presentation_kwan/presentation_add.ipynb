```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager

# 한글 폰트 설정 (Windows 환경)
font_path = "C:/Windows/Fonts/malgun.ttf"
font_prop = font_manager.FontProperties(fname=font_path)

# 데이터 로드
df = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/원자력발전소/국내원자력발전소현황.csv', encoding='EUC-KR')

# 날짜 컬럼을 datetime 형식으로 변환
df['상업운전개시일'] = pd.to_datetime(df['상업운전개시일'], errors='coerce')
df['설계수명만료일'] = pd.to_datetime(df['설계수명만료일'], errors='coerce')

# 연도만 추출하여 새로운 컬럼 추가
df['상업운전개시연도'] = df['상업운전개시일'].dt.year
df['설계수명만료연도'] = df['설계수명만료일'].dt.year

# 각 발전소별로 상업운전개시연도의 최솟값과 설계수명만료연도의 최댓값을 구함
df_range = df.groupby('발전소명').agg({'상업운전개시연도': 'min', '설계수명만료연도': 'max'}).reset_index()

# 그래프 시각화
plt.figure(figsize=(14, 8))

# 막대 그래프 생성: 각 발전소의 상업운전개시연도에서 설계수명만료연도까지를 하나의 막대로 표현
for i, row in df_range.iterrows():
    plt.barh(i, row['설계수명만료연도'] - row['상업운전개시연도'], left=row['상업운전개시연도'], color='skyblue', height=0.8)

# 발전소 이름을 y축에 추가
plt.yticks(range(len(df_range)), df_range['발전소명'], fontproperties=font_prop)
plt.xticks(fontsize=12, fontproperties=font_prop)

# 제목 및 레이블 설정
plt.title('발전소별 상업운전개시연도에서 설계수명만료연도까지의 기간 (평균)', fontsize=16, fontproperties=font_prop)
plt.xlabel('연도', fontsize=14, fontproperties=font_prop)
plt.ylabel('발전소명', fontsize=14, fontproperties=font_prop)

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 파일 읽기
df_nuclear_50 = pd.read_csv('C:/geo/Earthquake-Korea-DataAnalysis/dataset/최종데이터셋/대책현황/원자력발전소/국내원자력발전소_50km이내_지진발생현황.csv', encoding='EUC-KR')

# '발생년월일'에서 연도만 추출하고, 정수형으로 변환
df_nuclear_50['연도'] = pd.to_datetime(df_nuclear_50['발생년월일'], errors='coerce').dt.year.astype('Int64')  # .astype('Int64')로 정수형 변환

# 연도 범위 필터링 (2019-2023)
df_nuclear_50 = df_nuclear_50[df_nuclear_50['연도'].between(2019, 2023)]

# 거리 구간 설정
bins = [0, 10, 20, 30, 40, 50]
labels = ['0-10km', '10-20km', '20-30km', '30-40km', '40-50km']
df_nuclear_50['거리 구간'] = pd.cut(df_nuclear_50['거리(km)'], bins=bins, labels=labels)

# 연도 및 거리 구간별 지진 규모 평균 계산
distance_magnitude_year = df_nuclear_50.groupby(['연도', '거리 구간'])['규모'].mean().reset_index()

# 시각화
plt.figure(figsize=(14, 8))
sns.lineplot(data=distance_magnitude_year, x='연도', y='규모', hue='거리 구간', marker='o', linewidth=2)

# 그래프 꾸미기
plt.title('연도별 원전과의 거리 구간별 평균 지진 규모', fontsize=16)
plt.xlabel('연도', fontsize=14)
plt.ylabel('평균 지진 규모', fontsize=14)
plt.xticks(ticks=range(2019, 2024), labels=['2019', '2020', '2021', '2022', '2023'], fontsize=12)  # x축 연도 레이블 설정
plt.yticks(fontsize=12)
plt.legend(title='거리 구간', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 날짜에서 연도 추출
df_nuclear_50['발생년'] = pd.to_datetime(df_nuclear_50['발생년월일'], errors='coerce').dt.year

# 연도별 지진 발생 횟수 집계
yearly_earthquake_counts = df_nuclear_50.groupby('발생년').size()

# 연도별 원전 인근 가속도의 평균값 계산
yearly_avg_acceleration = df_nuclear_50.groupby('발생년')['원전인근가속도값(g)'].mean()

# 그래프 생성
fig, ax1 = plt.subplots(figsize=(10, 6))

# 첫 번째 축: 연도별 지진 발생 횟수
ax1.bar(yearly_earthquake_counts.index, yearly_earthquake_counts.values, color='skyblue', alpha=0.7, label='지진 발생 횟수')
ax1.set_xlabel('연도', fontsize=14)
ax1.set_ylabel('지진 발생 횟수', color='black', fontsize=14)
ax1.tick_params(axis='y', labelcolor='black')

# 두 번째 축: 연도별 평균 원전 인근 가속도 값
ax2 = ax1.twinx()  # 두 번째 y축을 생성
ax2.plot(yearly_avg_acceleration.index, yearly_avg_acceleration.values, color='salmon', marker='o', label='원전 인근 평균 가속도 값')
ax2.set_ylabel('평균 가속도 값 (g)', color='salmon', fontsize=14)
ax2.tick_params(axis='y', labelcolor='salmon')

# 그래프 제목 및 레이아웃 설정
plt.title('연도별 지진 발생 횟수 및 원전 인근 가속도 추이', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# 범례 설정: ax1 범례는 기본 위치, ax2 범례는 왼쪽에 설정
ax1.legend(loc='upper left')
ax2.legend(loc='upper left')

plt.tight_layout()  # 레이아웃을 자동으로 조정하여 겹치지 않도록 함
plt.show()
```

```python
# '내진설계기준'과 '내진성능향상값'을 숫자형으로 변환
df_nuclear_resist['내진설계기준'] = pd.to_numeric(df_nuclear_resist['내진설계기준'], errors='coerce')
df_nuclear_resist['내진성능향상값'] = pd.to_numeric(df_nuclear_resist['내진성능향상값'], errors='coerce')

# NaN 값 제거
df_nuclear_resist_clean = df_nuclear_resist.dropna(subset=['호기', '내진설계기준', '내진성능향상값'])

# 호기별 내진설계기준과 내진성능향상값 분석
analysis = df_nuclear_resist_clean.groupby('호기')[['내진설계기준', '내진성능향상값']].mean().reset_index()

# 시각화 (세로 막대 그래프)
plt.figure(figsize=(14, 8))

# 두 변수를 하나의 그래프에 표현
analysis_melted = analysis.melt(id_vars='호기', value_vars=['내진설계기준', '내진성능향상값'], 
                                var_name='항목', value_name='값')

# 세로 막대 그래프
sns.barplot(data=analysis_melted, x='호기', y='값', hue='항목', palette='coolwarm')

# 제목 및 레이블 설정
plt.title('호기별 내진설계기준과 내진성능향상값', fontsize=18, fontweight='bold')
plt.xlabel('호기', fontsize=14)
plt.ylabel('값', fontsize=14)

# 범례 설정
plt.legend(title='항목', fontsize=12)

# X, Y 축 레이블 설정
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# 레이아웃 최적화
plt.tight_layout()

# 그래프 표시
plt.show()
plt.show()
```
