### 1. 전 세계에서 발생한 지진 현황

<태블로에서 세계지도에 발생한 지진들을 표시하여 시각화>

지진은 판의 충돌, 지각의 변형, 화산 활동, 지하수 침반, 태풍으로 인한 지하수 수위 상승과 같은 원인 이외에도 인간의 지반 공사, 핵실험 등 다양한 이유로 발생하고 있습니다. 지진이 위험한 이유는 대규모 지반진동 뿐만 아니라 해일 동반, 낙상사고, 여진 등으로 인류에게 막대한 피해를 끼치기 때문입니다.

비교적 최근인 2004년 푸켓의 지진해일, 2005년 카슈미르 지진, 2008년 쓰촨성 지진, 2010년 아이티 지진,2011년에 일어난 일본 지진해일은 인류에게 막대한 피해를 주었습니다. 뿐만 아니라 2011년에 일어난 일본 지진해일을 통해서 원자력 발전소의 붕괴와 같은 대참사를 통해 지진의 위험성에 대해서 체감할 수 있었습니다.

그렇다면 과연 우리가 살고 있는 대한민국은 지진의 위협으로 부터 안전할까?라는 의문이 들었고, 이러한 의문을 바탕으로 데이터 분석을 시작하였습니다.

### 2. 대한민국 국민들의 자연재해 인식 현황

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
df_edu = pd.read_excel('..\찐최종 데이터셋\대책현황\교육\국내자연재해_국민인식현황_찐최종.xlsx')
df_edu
```

```python
df_edu.info()
```

```python
import pandas as pd

# CSV 파일 읽기

# '전체' '계' 행 추출
df_filtered = df_edu[(df_edu['특성별(1)'] == '전체') & (df_edu['특성별(2)'] == '계')]

# 연도별 데이터 추출 및 재구성
years = ['2008', '2010', '2012', '2014', '2016', '2018', '2020', '2022', '2024']
result_data = []

for year in years:
    row_data = [
        df_filtered[year].values[0],
        df_filtered[f'{year}.1'].values[0],
        df_filtered[f'{year}.2'].values[0],
        df_filtered[f'{year}.3'].values[0],
        df_filtered[f'{year}.4'].values[0]
    ]
    result_data.append([year] + row_data)

# 결과 데이터프레임 생성
df_result = pd.DataFrame(result_data, columns=['연도', '매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음'])
df_result = df_result.set_index('연도')

# 결과 출력
print(df_result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import numpy as np

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# CSV 파일 읽기

# '전체' '계' 행 추출
df_filtered = df_edu[(df_edu['특성별(1)'] == '전체') & (df_edu['특성별(2)'] == '계')]

# 연도별 데이터 추출 및 재구성
years = ['2008', '2010', '2012', '2014', '2016', '2018', '2020', '2022', '2024']
result_data = []

for year in years:
    row_data = [
        float(df_filtered[year].values[0]),
        float(df_filtered[f'{year}.1'].values[0]),
        float(df_filtered[f'{year}.2'].values[0]),
        float(df_filtered[f'{year}.3'].values[0]),
        float(df_filtered[f'{year}.4'].values[0])
    ]
    result_data.append([year] + row_data)

# 결과 데이터프레임 생성
df_result = pd.DataFrame(result_data, columns=['연도', '매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음'])
df_result = df_result.set_index('연도')

# 선 그래프 생성
plt.figure(figsize=(12, 6))
plt.title('연도별 국내 자연재해 안전도 인식 변화', fontsize=16)

# 눈에 띄는 다른 색상으로 변경
colors = ['#FF1493', '#1E90FF', '#32CD32', '#FFA500', '#8B008B']
categories = ['매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음']

for i, category in enumerate(categories):
    plt.plot(df_result.index, df_result[category], label=category, color=colors[i], marker='o', linewidth=2)

plt.xlabel('연도')
plt.ylabel('비율 (%)')
plt.legend(title='안전도 인식', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```
