### 지진이 발생하는 이유
지진은 판의 충돌, 지각의 변형, 화산 활동, 지하수 침반, 태풍으로 인한 지하수 수위 상승과 같은 원인 이외에도 인간의 지반 공사, 핵실험 등 다양한 이유로 발생하고 있습니다. 지진이 위험한 이유는 대규모 지반진동 뿐만 아니라 해일 동반, 낙상사고, 여진 등으로 인류에게 막대한 피해를 끼치기 때문입니다.

비교적 최근인 2004년 푸켓의 지진해일, 2005년 카슈미르 지진, 2008년 쓰촨성 지진, 2010년 아이티 지진,2011년에 일어난 일본 지진해일은 인류에게 막대한 피해를 주었습니다. 뿐만 아니라 2011년에 일어난 일본 지진해일을 통해서 원자력 발전소의 붕괴와 같은 대참사를 통해 지진의 위험성에 대해서 체감할 수 있었습니다.

그렇다면 과연 우리가 살고 있는 대한민국은 지진의 위협으로 부터 안전할까?라는 의문이 들었고, 이러한 의문을 바탕으로 데이터 분석을 시작하였습니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc


# 연령대 리스트
age_groups = ['13∼19세', '20∼29세', '30∼39세', '40∼49세', '50∼59세', '60∼69세', '70∼79세', '80세이상']

# 연도별 데이터 추출 및 재구성
years = ['2008', '2010', '2012', '2014', '2016', '2018', '2020', '2022']
result_data = []

def safe_float(x):
    try:
        return float(x)
    except ValueError:
        return np.nan

for age_group in age_groups:
    df_filtered = df_edu[(df_edu['특성별(1)'] == '연령') & (df_edu['특성별(2)'] == age_group)]
    
    for year in years:
        row_data = [
            safe_float(df_filtered[year].values[0]),
            safe_float(df_filtered[f'{year}.1'].values[0]),
            safe_float(df_filtered[f'{year}.2'].values[0]),
            safe_float(df_filtered[f'{year}.3'].values[0]),
            safe_float(df_filtered[f'{year}.4'].values[0])
        ]
        result_data.append([age_group, year] + row_data)

# 결과 데이터프레임 생성
df_result = pd.DataFrame(result_data, columns=['연령대', '연도', '매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음'])

# 연도별로 평균과 표준편차 계산
mean_std_data = []

for year in years:
    year_data = df_result[df_result['연도'] == year][['매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음']]
    means = year_data.mean()
    stds = year_data.std()
    mean_std_data.append([year] + list(means) + list(stds))

# 결과 데이터프레임 생성 (평균과 표준편차)
mean_std_columns = ['연도'] + [f'{col}_mean' for col in ['매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음']] + \
                   [f'{col}_std' for col in ['매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음']]
df_mean_std = pd.DataFrame(mean_std_data, columns=mean_std_columns)

# 막대 그래프 생성 (평균과 표준편차)
categories = ['매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음']
fig, axs = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle('2008년부터 2022년까지 연령대별 국내 자연재해 인식의 평균과 표준편차')

for i, category in enumerate(categories):
    row = i // 3
    col = i % 3
    
    means_col = f'{category}_mean'
    stds_col = f'{category}_std'
    
    x_pos = np.arange(len(years))
    
    # 막대 그래프 그리기
    axs[row, col].bar(x_pos, df_mean_std[means_col], yerr=df_mean_std[stds_col], capsize=5, alpha=0.7, color='skyblue')
    
    # 그래프 제목 및 라벨 설정
    axs[row, col].set_title(category)
    axs[row, col].set_xticks(x_pos)
    axs[row, col].set_xticklabels(years)
    axs[row, col].set_ylabel('비율 (%)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc


# 산업 리스트
industries = ['농림어업', '광공업', '사회간접자본․기타서비스업']

# 연도별 데이터 추출 및 재구성
years = ['2008', '2010', '2012', '2014', '2016', '2018', '2020', '2022']
result_data = []

def safe_float(x):
    try:
        return float(x)
    except ValueError:
        return np.nan

for industry in industries:
    df_filtered = df_edu[(df_edu['특성별(1)'] == '산업') & (df_edu['특성별(2)'] == industry)]
    
    for year in years:
        row_data = [
            safe_float(df_filtered[year].values[0]),
            safe_float(df_filtered[f'{year}.1'].values[0]),
            safe_float(df_filtered[f'{year}.2'].values[0]),
            safe_float(df_filtered[f'{year}.3'].values[0]),
            safe_float(df_filtered[f'{year}.4'].values[0])
        ]
        result_data.append([industry, year] + row_data)

# 결과 데이터프레임 생성
df_result = pd.DataFrame(result_data, columns=['산업', '연도', '매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음'])

# 산업별로 평균과 표준편차 계산
mean_std_data = []

for industry in industries:
    industry_data = df_result[df_result['산업'] == industry][['매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음']]
    means = industry_data.mean()
    stds = industry_data.std()
    mean_std_data.append([industry] + list(means) + list(stds))

# 결과 데이터프레임 생성 (평균과 표준편차)
mean_std_columns = ['산업'] + [f'{col}_mean' for col in ['매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음']] + \
                   [f'{col}_std' for col in ['매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음']]
df_mean_std = pd.DataFrame(mean_std_data, columns=mean_std_columns)

# 막대 그래프 생성 (평균과 표준편차)
categories = ['매우 안전', '비교적 안전', '보통', '비교적 안전하지 않음', '매우 안전하지 않음']
fig, axs = plt.subplots(3, 2, figsize=(20, 15))
fig.suptitle('2008년부터 2022년까지 산업별 국내 자연재해 인식의 평균과 표준편차')

for i, category in enumerate(categories):
    row = i // 2
    col = i % 2
    
    means_col = f'{category}_mean'
    stds_col = f'{category}_std'
    
    x_pos = np.arange(len(industries))
    
    # 막대 그래프 그리기
    axs[row, col].bar(x_pos, df_mean_std[means_col], yerr=df_mean_std[stds_col], capsize=5, alpha=0.7, color='skyblue')
    
    # 그래프 제목 및 라벨 설정
    axs[row, col].set_title(category)
    axs[row, col].set_xticks(x_pos)
    axs[row, col].set_xticklabels(industries)
    axs[row, col].set_ylabel('비율 (%)')

# 마지막 빈 서브플롯 제거
fig.delaxes(axs[2][1])

plt.tight_layout()
plt.show()
```
