```python
import matplotlib.pyplot as plt
import numpy as np

# 예제 데이터 생성 (필요시 사용)
eq = np.random.uniform(-8, 8, 100)  # 임의의 깊이 데이터 생성

# 차원 조정 (필요시)
if eq.ndim > 1:
    eq = eq.flatten()

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

fig, ax = plt.subplots()

# 박스플롯 그리기
ax.boxplot(eq)  # 리스트나 1차원 배열로 전달
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel("깊이(km)")

plt.show()
```

국내원자력발전소_내진설계현황

```python
df_sd = pd.read_csv('C:/Users/campus4D018/Desktop/TP/Earthquake-Korea-DataAnalysis/dataset/대책현황/원자력발전소/국내원자력발전소_내진설계현황.csv',encoding= 'euc-kr')
df_sd
```

```python
df_sd.info()
```

```python
df_sd['내진설계기준'] = df_sd['내진설계기준'].str.replace('g', '', regex=False)
df_sd['내진성능향상값'] = df_sd['내진성능향상값'].str.replace('g', '', regex=False)
```

```python
df_sd
```

```python
df_sd['내진설계기준'] = pd.to_numeric(df_sd['내진설계기준'], errors='coerce')
df_sd['내진성능향상값'] = pd.to_numeric(df_sd['내진성능향상값'], errors='coerce')
```
